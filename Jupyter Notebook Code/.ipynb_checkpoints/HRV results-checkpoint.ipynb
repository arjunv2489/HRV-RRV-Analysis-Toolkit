```python
import pandas as pd
df = pd.read_csv('HRVcleaned.csv')
```

# BPM

```python
import numpy as np
bpm = np.array(df["BPM"].dropna())
bpm_x = np.arange(len(bpm))

X = []
for i in range(len(bpm)):
    bpm_arr = [bpm_x[i], bpm[i]]
    X.append(bpm_arr)
```

```python
# import plotly.express as px
import plotly.graph_objects as go
import numpy as np
# fig = px.scatter(x=bpm_x, y=bpm,name='BPM')
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=bpm_x, y=bpm,
    name='BPM',
    mode='markers',
    marker_color='rgba(152, 0, 0, .8)'
))


fig.update_traces(mode='markers', marker_line_width=2, marker_size=10)
# fig.update_layout( )

fig.update_layout(
    yaxis=dict(
        tickmode = 'linear',
        tick0 = 35,
        dtick = 5
    ),
    xaxis_title="Patient ID",
    yaxis_title="BPM value",
    
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    ),
    title={
        'text': "BPM",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
    
)

# fig.update_yaxes(tickvals=np.arange(35, 130, 5))




# fig.update_yaxes(ticks="outside", tickwidth=2, tickcolor='crimson', ticklen=10, col=1)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd


X_axis = df["BPM"].dropna()
Y = np.arange(len(X))

X = np.array(df["BPM"].dropna()).reshape(-1, 1) 
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
labels = kmeans.labels_


val = np.array(df["BPM"].dropna())
val_x = np.arange(len(bpm))

# p1 = []
# for i in range(len(val)):
#     if
#     arr = [val_x[i], val[i],labels[i]]
#     abc.append(arr)

efg = pd.read_csv('HRV Data - Sheet1.csv')
efg = efg['Scores'].dropna().values.tolist()
p1 = []
p2 = []
score_value = 0
for i in range(len(val)):
    if labels[i] == 0:
        if efg[i] > 8 :
            arr = [val_x[i], val[i],labels[i],0]
            score_value += 1
            p1.append(arr)
        else:
            arr = [val_x[i], val[i],labels[i],1]
            p1.append(arr)
    else:
        if efg[i] <= 8 :
            arr = [val_x[i], val[i],labels[i],1]
            score_value += 1
            p2.append(arr)
        else:
            arr = [val_x[i], val[i],labels[i],0]
            p2.append(arr)


df_sdnn_0 = pd.DataFrame(data=p1, columns=['No', 'SDNN', 'Predicted Label', 'True Label'])
df_sdnn_1 = pd.DataFrame(data=p2, columns=['No', 'SDNN', 'Predicted Label', 'True Label'])    

# fig = px.scatter(x=bpm_x, y=bpm,name='BPM')
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_sdnn_0['No'], y=df_sdnn_0['SDNN'],
    name='Anxiety',
    mode='markers',
    marker_color='rgba(125, 0, 0, .8)'
))
fig.add_trace(go.Scatter(
    x=df_sdnn_1['No'], y=df_sdnn_1['SDNN'],
    name='No Anxiety',
    mode='markers',
    marker_color='rgba(23,67,88,0.5)'
))

fig.update_traces(mode='markers', marker_line_width=2, marker_size=10)
# fig.update_layout( )

fig.update_layout(
    
    xaxis_title="Patient ID",
    yaxis_title="BPM value",
    
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    ),
    title={
        'text': "BPM K-MEANS Cluster",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
    
)
```

```python
sdnn_df = df_sdnn_0.append(df_sdnn_1, ignore_index = True) 
print("Score: " ,(score_value/len(sdnn_df))*100)
```
